```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing the Data

```python
iris_df=pd.read_csv('C:/Users/krite/Desktop/Iris-Dataset.csv')
iris_df=iris_df.iloc[:,1:]
iris_df.head()
```

Preparing the train and test datasets

```python
X=iris_df.iloc[:,:4].values
y=iris_df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=0)
```

## Defining decision tree classifier model and fitting on train data

```python
from sklearn.tree import DecisionTreeClassifier

dtree=DecisionTreeClassifier()
dtree.fit(X_train,y_train)

print('Decision Tree Classifer has been Created')
```

## Prediction on test data accuracy score

```python
dtree.predict(X_test)
```

```python
y_test
```

```python
dtree.score(X_test,y_test)
```

The accuracy of this decision tree model was 96.67%

## Visualizing the Decision Tree Classifier Graphically

```python
from six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

# Visualize the graph
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data, feature_names=iris_df.iloc[:,:4].columns,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
